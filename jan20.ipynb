```python
import pandas as pd
```

```python
df = pd.read_csv('data/zoo.csv',
                 encoding='latin1')

df.head()
```

```python
df.info()
```

```python
# see if there are duplicates
duplicateAnimals = df["animal"].value_counts()
duplicateAnimals = duplicateAnimals[duplicateAnimals > 1]

df[df["animal"].isin(duplicateAnimals.index)]
```

```python
df.groupby("animal").size().sort_values(ascending=False)
```

```python
dfSingle = df.drop_duplicates(subset="animal", keep="first")
dfSingle.groupby("animal").size().sort_values(ascending=False)
```
